```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from bs4 import BeautifulSoup
import requests
```

```python
header = {'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) '\
           'AppleWebKit/537.36 (KHTML, like Gecko) '\
           'Chrome/75.0.3770.80 Safari/537.36'}
resp = requests.get("https://www.bestbuy.com/site/playstation-5/ps5-games/pcmcat1587395108347.c?id=pcmcat1587395108347", headers = header)
resp
```

```python
soup = BeautifulSoup(resp.text, 'html5lib')
soup
```

```python
items = soup.find(class_ = "sku-item-list")
items
```

```python
each_item = items.find_all(class_ = "sku-item")
each_item
```

```python
prices = []
for j in each_item:
    price_block = j.find(class_ = "price-block")
    price = price_block.find(class_ = "sku-list-item-price")
    actual_price = price.find(class_ = "priceView-hero-price priceView-customer-price")
    if actual_price != None:
        #print(actual_price.get_text())
        price = actual_price.find({"span": "aria-hidden"})
        #print(price.get_text())
        prices.append(price.get_text())
prices
```

```python
game_titles = []
for j in each_item:
    title = j.find(class_ = "sku-title")
    #print(title)
    game_title = title.find(class_ = "sku-header")
    
    #actual_price = price.find(class_ = "priceView-hero-price priceView-customer-price")
    if game_title != None:
        #print(actual_price.get_text())
        #price = actual_price.find({"span": "aria-hidden"})
        #print(price.get_text())
        game_titles.append(game_title.get_text())
game_titles
```

```python
game_ratings = []
for j in each_item:
    rating = j.find(class_ = "variation-info")
    game_rating = rating.find(class_ = "sku-model").find_next("div").find_next("div").find_next("div").find_next("div")
    #print(game_rating.get_text())
    game_game_rating = game_rating.a.get_text()
    #print(game_game_rating)
    game_ratings.append(game_game_rating)
    #game_ratings.append(rating.find_next("div").find_next("div").find_next("div").find_next("div").find_next("div").find_next("a").get_text())
#game_ratings
game_ratings
```

```python
publishers = []
for j in each_item:
    publisher = j.find(class_ = "variation-info")
    publisher_publisher = publisher.find(class_ = "attribute-title")
    publisher_publisher_publisher = publisher.find(class_ = "sku-value")
    publishers.append(publisher_publisher.get_text() + " " + publisher_publisher_publisher.get_text())
    #publishers.append(rating.find_next("div").find_next("div"))
publishers
```

```python
dates = []
for j in each_item:
    date = j.find(class_ = "variation-info")
    #publisher_publisher = publisher.find(class_ = "attribute-title")
    #publisher_publisher_publisher = publisher.find(class_ = "sku-value")
    new_date = date.find(class_ = "sku-model").find_next("div").find_next("div").find_next("div")
    #print(new_date.get_text())
    dates.append(new_date.get_text())
    #dates.append(date.find_next("div").find_next("div").find_next("div").find_next("div").get_text())
    #publishers.append(rating.find_next("div").find_next("div"))
dates
```

```python
reviews = []
for j in each_item:
    review = j.find(class_ = "variation-info")
    #publisher_publisher = publisher.find(class_ = "attribute-title")
    #publisher_publisher_publisher = publisher.find(class_ = "sku-value")
    review_rating = review.find(class_ = "ugc-ratings-reviews c-ratings-reviews-v4 c-ratings-reviews-v4-size-small")
    reviews.append(review_rating.get_text())
    #publishers.append(rating.find_next("div").find_next("div"))
#dates
reviews
```
